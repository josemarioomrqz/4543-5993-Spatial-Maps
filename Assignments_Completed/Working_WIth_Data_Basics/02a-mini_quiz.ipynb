# Mini Quiz

### 1️⃣ Quiz Question - Given the python list: 

```python
L = [13,17,31,37,73]
```
Write one line of Python code that removes the value 31 from the list by value, not by index.
After your code executes, the list should become:

```python
L = [13,17,31,37,73]
```

````python
# Answer 1
L = [13,17,31,37,73]
L.remove(31)
print(L)

````

### 2️⃣ Quiz Question — Accessing Tuple Elements

Given the following tuple:

```python
T = (13, 17, 31, 37, 73)
```

Write one line of Python code that accesses the second element of the tuple and stores it in a variable named x.

````python
# Answer 2
T = (13,17,31,37,73)
x = T[1]
print(x)
````

### 3️⃣ Quiz Question — Changing Tuple Elements

Given the following tuple:

```python
T = (13, 17, 31, 37, 73)
```

Write the necessary Python code that would change the second element of the tuple to 11.

````python
# Answer 3
T = (13,17,31,37,73)
T = (T[0], 11, T[2], T[3], T[4])
print(T)

````

### 4️⃣ Quiz Question — Dictionary keys

Given the following dicitonary:

```python
D = {
    "apple":"10",
    "banana":"11",
    "cherry":"8"
}
```

Write the necessary Python code that would print the keys of the the dictionary D.

````python
# Answer 4
D = {
    "apple":"10",
    "banana":"11",
    "cherry":"8"
}
for key in D:
    print(key)


````
